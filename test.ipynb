```python
import gym
import babyai_text
import babyai.utils

env = gym.make('BabyAI-MixedTrainLocal-v0')
env.seed(100)
counter = 0


observation, info = env.reset()
print(env.action_space)

while(True):
  action = env.action_space.sample()
  observation, reward, done, info = env.step(action)
  
  counter += 1
  print(counter)
    
  babyai.utils.viz(env)
  print(babyai.utils.info(reward, heading="Reward"))
  print(babyai.utils.info(info, "Subtasks"))
        
  if done:
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
    
    counter += 1
    print(counter)
    
    babyai.utils.viz(env)
    print(babyai.utils.info(reward, heading="Reward"))
    print(babyai.utils.info(info, "Subtasks"))
    env.close()
    break
```
